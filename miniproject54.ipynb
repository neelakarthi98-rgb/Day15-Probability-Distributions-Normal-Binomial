```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Set average rate (lambda)
lam = 3

# Range of possible events (number of crashes)
x = np.arange(0, 11)

# PMF (probability of exactly k crashes)
pmf_values = poisson.pmf(x, lam)

# CDF (probability of k or fewer crashes)
cdf_values = poisson.cdf(x, lam)

# Plot PMF and CDF side by side
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# --- PMF Plot ---
ax[0].bar(x, pmf_values, color='skyblue')
ax[0].set_title('Poisson PMF (λ = 3)\nProbability of exactly k crashes')
ax[0].set_xlabel('Number of crashes (k)')
ax[0].set_ylabel('P(X = k)')
ax[0].set_xticks(x)

# --- CDF Plot ---
ax[1].step(x, cdf_values, where='mid', color='orange')
ax[1].set_title('Poisson CDF (λ = 3)\nProbability of k or fewer crashes')
ax[1].set_xlabel('Number of crashes (k)')
ax[1].set_ylabel('P(X ≤ k)')
ax[1].set_xticks(x)

plt.tight_layout()
plt.show()
```
