//////###Simulating and Visualizing a Poisson Distribution####////

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 

lambda_customers=5
interval=1000

arrivals=np.random.poisson(lam=lambda_customers,size=interval)

plt.figure(figsize=(8,8))
sns.histplot(arrivals,bins=range(0,max(arrivals)+1),color="pink")
plt.xlabel("number of customers")
plt.ylabel("frequency")
plt.xticks(range(0,max(arrivals)+1))
plt.title("Distribution of Customer Arrivals at Coffee Shop")
plt.tight_layout()
plt.show()
```
