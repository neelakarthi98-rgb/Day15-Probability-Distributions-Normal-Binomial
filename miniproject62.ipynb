## Simple PCA for Image Compression

////Objective: Use PCA to compress a simple grayscale image.///

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread

# 1 Generate a simple grayscale image (100x100)
np.random.seed(42)
original_image =imread("nature.jpg")# values between 0 and 1
if original_image.ndim==3:
    image_gray=original_image[:,:,0]+original_image[:,:,1]*0.5870+original_image[:,:,2]*0.1140
else:
    image_gray=original_image
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(original_image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f"Grey image")
plt.imshow(image_gray, cmap='gray')
plt.axis('off')

plt.show()    
```
