//////Generating and Analyzing a Normal Distribution////

******Generate a random dataset that follows a normal distribution and verify its properties****

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate Data
mu = 170        # Mean
sigma = 5       # Standard deviation
n = 1000        # Number of data points

data = np.random.normal(mu, sigma, n)
```

//// Statistical Analysis////

```python
mean_data = np.mean(data)
std_data = np.std(data)
```

/////Empirical Rule///

```python
within_1_std = np.sum((data >= mean_data - std_data) & (data <= mean_data + std_data))
within_2_std = np.sum((data >= mean_data - 2*std_data) & (data <= mean_data + 2*std_data))
within_3_std = np.sum((data >= mean_data - 3*std_data) & (data <= mean_data + 3*std_data))

percent_1_std = (within_1_std / n) * 100
percent_2_std = (within_2_std / n) * 100
percent_3_std = (within_3_std / n) * 100

print(f"Calculated Mean: {mean_data:.2f}")
print(f"Calculated Standard Deviation: {std_data:.2f}")
print(f"Percentage of data within 1 std: {percent_1_std:.2f}%")
print(f"Percentage of data within 2 std: {percent_2_std:.2f}%")
print(f"Percentage of data within 3 std: {percent_3_std:.2f}%")
```

///// Visualization////

```python
plt.figure(figsize=(10,6))
sns.histplot(data, bins=30,kde=True, color='orange')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
