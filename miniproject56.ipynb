```python
import numpy as np
import matplotlib.pyplot as plt
data=np.random.normal(100,10,20)
def caluculate_z_score(x,mean,std_dev):
    z=(x-mean)/std_dev
    return z
mean=np.mean(data)
x=100
std_dev=np.std(data)
z_score=caluculate_z_score(x,mean,std_dev)
print(data)
print("")
print(f"mean value is:{mean:.2f}")
print(f"standared deviation of the value is:{std_dev:.2f}")

print(f"z score is:{z_score:.2f}")
plt.figure(figsize=(10,5))

plt.plot(data, color='red', linestyle='--',label="mean")
plt.xlabel("Data values")
plt.ylabel("Z-score")
plt.title("Z-scores of all values in the dataset")
plt.legend()
plt.show()
```
